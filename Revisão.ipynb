# MODELO REGRESSÃO LINEAR

O que é REGRESSÃO LINEAR? É o estudo da estatisitica que investiga e relação entre duas ou mais variáveis relacionadas de maneira deterministica.

Em outras palavras temos uma variável y cujo resultado está correlacionado com a variação do valor de uma variável independente x

Com base nisso temos o modelo de regressão linear simples

---



## MODELO DE REGRESÃO LINEAR SIMPLES
#$ Y = \beta_0 + \beta_1 . X$

# Y - variável resposta e dependente
# X - variável independente, preditora
# $\beta_0$ - intercepto
# $\beta_1$ - coeficiente angular

---



O medelo de regressão linear permite plotar os dados em um gráfico que representa a dispesão desses dados, quando os dados estão concentrados em torno de uma reta cujo coeficiente seja em torno de 45° temos o que chamamos dos dados bem ajustados ao modelo.

A correlação pode ser classificada como:

## POSITIVA - Quando a reta é ascendente, e o coeficiente de correlação varia de 0 a 1

## NEGATIVA - Quando a reta é descendente, e o coeficiente de correlação varia de 0 a -1

# MODELO PROBABILÍSTICO LINEAR

O modelo probalilístico linear leva em consideração um provável erro aleatório que pode influênciar o valor da variável resposta (y) para mais ou para menos em relação ao valor esperado.

O modelo nos permite predizer um determinado valor de y quando x assumir um valor n com base na na reta de regressão e no erro padrão através do cálculo da Probabilidade da Distribuição Normal

## MODELO PROBABILÍSTICO LINEAR
#$ Y = \beta_0 + \beta_1 . X$ + e (erro residual)

# COMO DETERMINAR OS PARÂMETROS DA RETA DE REGRESSÃO LINEAR?

Lembrando que os parametros da reta de regressão são o $\beta_0$ e $\beta_1$, podemos analisar os dados de entrada das variáveis **x** e **y** para determinar o valor dos parametros que definirão a reta de regressão.

Esse cálculo é realizado através do MÉTODO DOS MÍNIMOS QUADRADOS. Esse métodos permite através aplicar fórmulas especificas para determinar $\beta_0$ e $\beta_1$.

---



# COMO DETERMINAR O $\beta_1$ ?

Para estimar o valor de $\beta_1$ temos 2 formúlas, mas antes, porque começar com $\beta_1$ e não $\beta_0$. É interesse sempre calcular $\beta_1$ primeiro pois seu valor pode ser usado para encontrar o $\beta_0$.

---



## FORMULA 1:

$ \frac{Sum ( Xi - ~X) . ( Yi - ~Y)}{Sum ( Xi - ~X) ^2}$

obs:
sum() é a somatória de todas as entradas da variável
Xi ou Yi é cada entrada de x ou y e
$\hat{x}$ e $\hat{y}$ é o x e y médios

---



## FORMULA 2 MAIS FÁCIL: $\frac {Sxy} {Sxx}$

Onde:

###$ S xy =  \frac{Sum (Xi . Yi)}{1} - \frac {( Sum (Xi) . Sum (Yi) )} {n}$

ou $ n * Sum (Xi . Yi) - ( Sum (Xi) . Sum (Yi) )$

###$ S xx =  \frac{Sum (Xi ^2)}{1} - \frac { Sum (Xi) ^2 } {n}$

ou $ n * Sum (Xi^2) - ( Sum (Xi)^2)$

obs: n é o número total de amostras

---



# COMO DETERMINAR O $\beta_0$ ?

Para estimar o valor de $\beta_0$ também temos 2 formas:

## FORMULA 1:

#$\frac { Sum (Yi) - \beta_1 . Sum (Xi) } {n}$

obs:
sum() é a somatória de todas as entradas da variável
Xi ou Yi é cada entrada de x ou y
b1 já foi determinado anteriormente

---



## FORMULA 2:

#  $\hat{y} - \beta_1 . \hat{x}$


obs:
$\hat{x}$ e $\hat{y}$ é o x e y médios
b1 já foi determinado anteriormente






# EXEMPLO DE APLICAÇÃO

Considerando que temos os seguintes valores associados as variáveis x e y:

## x = (1, 2, 0)

## y = (3, 5, 1)

Determine os valores dos parametros $\beta_0$ e $\beta_1$ da provável reta de regressão linear

```python
# primeiramente vamos listar nossas variáveis:

x = [1, 2, 0]

y = [3, 5, 1]


# agora vamoos determinar alguns outros valores, como x², y² e x.y:

# vou utilizar o numpy para calcular o quadrado e produto das variáveis

import numpy as np

x2 = np.square(x)

y2 = np.square(y)

xy = np.multiply(x, y)


# agora vou colocar tudo em data frame para visualizar melhor

import pandas as pd

df = pd.DataFrame({"x": x, "y": y,"x2": x2, "y2": y2,",xy": xy})

# agora vou adcionar a somatória de cada coluna na ultima linha da tabela

# calculamos a soma de cada coluna
somas = df.sum(axis=0)

# adicionamos as somas como uma nova linha no dataframe
df = df.append(somas, ignore_index=True)

# imprimimos o dataframe
print(df)
```

Com essa tabela fica mais fácil calcular os parametros da reta regressão, basta apenas substituir os valores nas fórmulas

```python
# Calculando o b1

n = 3

Sxy = sum(xy) - sum(x) * sum(y) / n
Sxx = sum(x2) - sum(x) * sum(x) / n

b1 = Sxy/Sxx

print(f"valor do b1 é: {b1}")

# Calculando o b0


b0 = ( sum(y) - b1 * sum(x) ) / n

print(f"valor do b0 é: {b0}")

# Determinado a minha reta de regressão:

print(f"Reta é : Y = {b0} + {b1} * X")
```

# VARIANÇA E DESVIO PADRÃO

Outra medida importante é o cálculo da variança que é usada para identificar intervalos de confiança e previsão. Geralmente a variança é dada pela fórmula:

##S² = $\frac{Sum(Xi - \hat{x})^2}{n} $

Sendo o Desvio Padrão = $\sqrt{S^2}$

Na reta de regressão a variança deve ser obtida através da Soma do Quadrado dos Erros (SQE), que mede a distância residual entre cada valor observado de y e seu valor previsto, o SQE pode ser obtido com:

# SQE

##SQE = Sum(y observado - y previsto)²

Ou:  

##$Syy - \beta_1 * Sxy$

Onde Syy é igual ao valor de SQT:

##$ SQT = Syy = \frac{Sum(y^2)}{1} -  \frac{Sum(y)^2}{n}$

 ou $n *  Sum (y^2) - Sum (y)^2$

<br>

Portanto para calcular o SQE poderia realizar-se uma iteração para calcular o quadrado da diferença de y e ^y, ou de forma fatorada:

$ n^2 * Sum(y²) - Sum(y)^2 + Sum (x^2) - (Sum (x)^2 - n * (Sum (Xi . Yi) - ( Sum (Xi) . Sum (Yi)^2 ) $

<br>

Depois de encontrarmos o valor de SQE poderemos calcular a variança:

##V² = $\frac {SQE} {n-2}$

<br>

# SQT

Como já mencionado anteriormente, SQT, a soma dos quadrados totais é dada pelo calculo de Syy:

$ SQT = Syy = \frac{Sum(y^2)}{1} -  \frac{Sum(y)^2}{n}$

<br>

# SQR

SQR é a Soma dos Quadrados dos Resíduos, é dado pela diferença entre SQT e SQE representando o quanto de erro residual nosso modelo tem.

##SQR = SQT - SQE

<br>

# Coeficiente de Determinação R²

O coeficiente de determinação permite demonstrar se o modelo linear é eficiente para ajustar a correalação das variáveis à reta de regressão, ele pode ser calculado por:

$R2 = 1 - \frac {SQE} {SQT} $


# CALCULANDO O VALOR DE Y PREVISTO

Considerando um cenário onde as variáveis x e y representam:

x = Capacidade Volumétrica do motor de determinados carros

y = Consumo de combustível destes carros

A partir dos dados abaixo Vamos estimar passo a passo a reta de regressão e calcular o valor esperado de y para x = 3200.

```python
import numpy as np
import pandas as pd

x = [792,994,1000,1368,1598,1796,1997,1999,2996,3197,3498,5461,6162,7291,999,1199,1399,1498,1598,1798,1998,2995,3493,3799,3982,5204,5980,7993]

y = [12,10.4,12.8,10.3,10.5,8.5,9.5,9.2,6.5,7.1,6.6,4.2,6.4,3,11.4,11.2,11,11,10.6,8.7,10.5,7.8,7,5,6.1,2.6,3,3.8]


x2 = np.square(x)

y2 = np.square(y)

xy = np.multiply(x, y)

n = 28 # nmr amostras


# Calculando o b1

Sxy = sum(xy) - (sum(x) * sum(y)) / n
Sxx = sum(x2) - (sum(x) * sum(x)) / n

b1 = Sxy/Sxx

print(f"valor do b1 é: {b1}")

# Calculando o b0

b0 = ( sum(y) - b1 * sum(x) ) / n

print(f"valor do b0 é: {b0}")

# Determinado a minha reta de regressão:

print(f"Reta é : Y = {b0} + {b1} * X")
```

```python
# Calculando o valor de y previsto para x = 3

x_var = 3000
y_prev = 12.144010117932384 + -0.0013478460028317721 * x_var
print(f"O valor de y esperado para x=3 é: {y_prev}")
```

# CALCULANDO A VARIANÇA ESTIMADA E COEFICIENTE DE DETERMINAÇÃO

Agora vamos calcular a variança estimada, através do SQT e SQE. E coeficiente de determinação R2

```python
# Calculando primeiramente o SQT (Syy)

Syy = sum(y2) - ( sum(y) * sum(y) ) / n

print(f"valor do SQT é: {Syy}")


# Agora calculamos o SQE

SQE = Syy - (b1 * Sxy)

print(f"valor do SQE é: {SQE}")


# Depois de encontrarmos o valor de SQE poderemos calcular a variança:

V2 = SQE / (n-2)

print(f"valor da variança estimada é: {V2}")

# Calculando o coeficiente de determinação temos:

SQT = Syy
R2 = 1 - (SQE/SQT)

print(f"O valor do coeficiente de determinação é: {R2}")
```

# INFERÊNCIAS SOBRE O β1

Muita vezes realizamos a regressão linear em um data set mas os parâmetros  calculados não fazem com que os dados estejam ajustados corretamente em uma dispersão confiável em torno da reta de regressão, portanto é necessário fazermos inferências acerca do parâmetro  β1, coeficiente angular para verificar ele produz uma relação adequada nos resultados de y.


Para testarmos a significância de β1 para a regressão podemos calcular o intervalo de confiança para determinar em que margem podemos trabalhar com um bom parâmetro β1, dado pela fórmula:

$\hat{\beta_1} \pm t_{\frac{\alpha}{2},n-2} * s_{\hat{\beta_1}} $


- t é o valor da estatisitica de teste calculada pela tabela t-student
- a = 1- nível de significância (ex: 1-95% = 0,05)
- a/2 = 0,025 que é a coluna da tabela t-student
- n-2 é o grau de liberdade, n o numero de amostra (28-2) gl = 26
- $s_{\hat{\beta_1}}$ é erro padrão estimado para $\beta_1$, calculdo pela fórmula:

    - $s_{\hat{\beta_1}} = \frac{s}{\sqrt{Sxx}}  $

Sendo que "s" : desvio padrão (raiz da variança estimada)  

Vamos demonstrar o cálculo abaixo:

```python
# Calculando o intervalo de confiança do nosso b1 =  -0.0013478460028317721

# Iremos utilizar um nivel de significância de 95%

# Primeiro vamos calcular nosso Sb1

# Já temos alguns dados calculados anteriormente como:

varianca = s2 = 1.5763506403791108
#Sxx calculado anteriormente, Então:

Sb1 = np.sqrt(s2) / np.sqrt(Sxx)

print(f"O valor do erro padrão estimado Sb1 é: {Sb1}")

# Temos que n = 28, logo n-2 + 26
# a = 0,95, logo 1-a /2 = 0,025

# Então temos que encontrar o valor da estatistica t, chamado de T critico

# Podemos fazer isso olhando na tabela t-student ou usando um método do python
```

Para pesquisar na tabela t-student olhamos na coluna o valor do a/2 (0,025) e truncamos com a linha do valor do grau de liberdade n-2 (26).

# TESTE DE HIPÓTESE PARA $\beta_1$

Outro método que podemos usar para verificar a validade do modelo é o teste de hipóteses, que nos permite verificar a utilidade do modelo, como a hipotése em que o $\beta_1$ tem siginificância = 0, indicando que o modelo deve ser rejeitado.

Para realizar o teste de hipótese devemos:

###**Definir hipóteses nula (H0) e alternativa (Ha)**

$H_0: \beta_1 = 0$

$H_a: \beta_1 \neq 0$

Isso indica que se H0 for confirmar então devemos rejeitar o modelo, porém se a estatisitca de teste permitir rejeitarmos H0 então confirmamos Ha e validamos o modelo.

###**Definir o Nível de Siginificância e Teste de Estatística**

Para um nível de significância de 95% temos $\alpha=0,05$

- definimos o $t_{crit}$ para o n-2=16 gl e $\alpha/2=0,025$

- definimos a estatística de teste sendo que

- $t=\frac{\hat{\beta}_1-\beta_{10}}{s_{\hat{\beta}_1}}$

Quando nosso H0 for = 0, podemos utilizar:

- $t=\frac{\hat{\beta}_1}{s_{\hat{\beta}_1}}$

###**Realizar o teste de hipótese**

Agora é necessário comparar se:

$|t| \geq t_{crit}$

Como no nosso exemplo teriamos:

$ |-11.39| \geq 2.05$

Como o teste de hipótese retorna verdadeiro, temos evidências suficientes para rejeitar $H_0$ em favor de $H_a$,  $H_a: \beta_1 \neq 0$ é verdadeira.

Dito de outra forma temos evidências suficientes para inferir que nosso $\beta_1$ é verdadeiro


```python
# Realizando Teste de Hipótese com os valores obtidos anteriormente

t = b1 / Sb1

# Teste de validação para retornar modulo no caso de t negativo
if t < 0:
    t = t * -1

# Teste de Hipótese

Res = t >= tcrit

if Res == True:
    print("Devemos rejeitar H0 e assumir H1")
else:
    print("Sem evidência suficiente para rejeitar H0")
```

# INTERVALO DE CONFIANÇA SOBRE O VALOR ESPERADO DE Y

Através de uma determinada reta de regressão dada por $ Y = \beta_0 + \beta_1 . x^*$ em que $x^*$ é algum valor determinado como entrada de x, podemos encontrar o valor esperado de y ($\hat{y}$):

$\mu_{Y,x^*}=\beta_0 + \beta_1 x^*$

<br>

Assim a variança de $\hat{y}$ pode ser obtida por:

$s_{\hat{Y}}=s\sqrt(\frac{1}{n}+\frac{(x^* -\bar{x})^2}{S_{XX}})$


<br>

## INTERVALO DE CONFIANÇA DE Y

Com esses dados podemos calcular o Intervalor de Confiança para o valor de entrada de $x^*$


IC de $\mu_{Y,x^*}$ pode se calcular com $\hat{Y}_1 \pm t_{\frac{\alpha}{2},n-2} s_{\hat{Y}} $



Observe que o valor de $s_{\hat{Y}}$  varia para cada valor de X porque depende do $x^*$

Assim precisa calcular o $\mu_{Y,x^*}$ e  o$s_{\hat{Y}}$ para cada amostra de x.

<br>

**Cálculos**

Vamos usar nossos dados para calcular o $\mu_{Y,x^=3300}$ e seu intervalo de confiança para um nível de significância de 95%.



```python
# Calculando o valor esperado de y para x = 3300

xi = 3300

y_prev = 12.144010117932384 + -0.0013478460028317721 * xi

# Agora calculamos a variança de y esperado

Sy_prev = s * (np.sqrt((1/n)+(((xi-xMed)*(xi-xMed))/(Sxx))))

# Agoraca Calculamos o Intervalo de Confiança

IC_1 = y_prev - tcrit * Sy_prev
IC_2 = y_prev + tcrit * Sy_prev

print("O intervalo de confiança de y para x=3300 vai de {0} a {1}".format(IC_1,IC_2))
```

# INTERVALO DE PREVISÃO DE Y

Para calcular o intervalo de previsão de um $\hat{y}$ temos apenas algumas mudanças em relação a fórmula do intervalo de confiança:

IP de $\mu_{Y,x^*}$ pode se calcular com $\hat{Y}_1 \pm t_{\frac{\alpha}{2},n-2} * \sqrt {s^2 + s^2_{\hat{Y}}} $

Da mesma forma o  o valor de $s_{\hat{Y}}$  varia para cada valor de X porque depende do $x^*$

Assim precisa-se calcular o $\mu_{Y,x^*}$ e  o$s_{\hat{Y}}$ para cada amostra de x.

<br>

**Cálculos**

Vamos usar nossos dados para calcular o IP de $\mu_{Y,x^=3300}$ para um nível de significância de 95%.




```python
# Calculando o valor esperado de y para x = 3300

Xi = 3300

Y_prev = 12.144010117932384 + -0.0013478460028317721 * Xi

# Agora calculamos a variança de y esperado

SY_prev = s * (np.sqrt((1/n)+(((Xi-xMed)*(Xi-xMed))/(Sxx))))
SY_prev2 = SY_prev * SY_prev
# Agoraca Calculamos o Intervalo de Confiança

ic_1 = Y_prev - tcrit * np.sqrt(s2 + SY_prev2)
ic_2 = Y_prev + tcrit * np.sqrt(s2 + SY_prev2)

print("O intervalo de previsão de y para x=3300 vai de {0} a {1}".format(ic_1,ic_2))
```

# DEFININDO FUNÇÃO PARA OBTER A RETA DE REGRESSÃO

Podemos também criar uma função para realizar o cálculo da reta de regressão, e depois só forncer as variáveis para obter o resultado da equação e demais estatisiticas.

```python
#@title função linreg
#O seguinte código pode também ser usado previamente para calcular SQE, SQT, SQR e r2 (Alternativa 2)
# fonte: Adaptado de https://www.geeksforgeeks.org/sum-function-python/
def linreg(X, Y):
    """
    retorna a solução b0 e b1 para a reta y = b0 + b1x e todas as somatórias necessárias
    """
    N = len(X)
    Sx = Sy = Sxx = Syy = Sxy = 0.0
    for x, y in zip(X, Y):
        Sx = Sx + x           # Somatória do termos x
        Sy = Sy + y           # Somatória do termos y
        Sxx = Sxx + x*x       # Somatória do termos x^2
        Syy = Syy + y*y       # Somatória do termos y^2
        Sxy = Sxy + x*y       # Somatória do termos x vezes y
    det = Sxx * N - Sx * Sx   # variável auxiliar
    Sxxt= Sxx - Sx*Sx/N       # Termo Sxx
    Sxyt= Sxy - Sx*Sy/N       # Termo Sxy
    Syyt= Syy - Sy*Sy/N       # Termo Syy
    r=Sxyt/pow((Sxxt*Syyt), 1/2)
    return (Sxy * N - Sy * Sx)/det, (Sxx * Sy - Sx * Sxy)/det, Sx, Sy, Sxx,  Syy , Sxy, Sxxt, Sxyt, Syyt,r
```

```python
#obter os valores necessários para calcular SQE e afins
x=df['x']
y=df['y']
b1,b0,sox,soy,sox2,soy2,soxy,sxx,sxy,syy,corr = linreg(x,y)
#Calcular SQE, SQT e SQR
sqe=syy-b1*sxy
sqt=syy
sqr=sqt-sqe
#calcular r^2 coeficiente de determinação
r2=1-sqe/sqt
#Calcular variância estimada
s2=sqe/(len(y)-2)
#Calcular desvio padrão estimado
s=pow(s2,1/2)
#Calcular erro padrão
ebeta=s/pow(sxx,1/2)

print('b0= {0},b1= {1}'.format(b0,b1))
print('Somatória x= {0}, Somatória y= {1},Somatória xy= {2}'.format(sox,soy,soxy))
print('Somatória x^2= {0}, Somatória y^2= {1}'.format(sox2,soy2))
print('Sxx= {0}, Syy {1}, Sxy {2}'.format(sxx,syy,sxy))
print('Coeficiente de correlação= {0}'.format(corr))
print('SQE= {0}, SQT {1}, SQR {2}'.format(sqe,sqt,sqr))
print('Variância estimada, s^2= {0}, Desvio padrão estimado, s= {1}'.format(s2,s))
print('Coeficente de determinação, r^2= {0}, Erro padrão, s_b1= {1}'.format(r2,ebeta))
```

# VERIFICANDO A ADEQUABILIDADE DO MODELO

É necessário verificar a adequabilidade de modelos de regressão linear para garantir que o modelo esteja descrevendo corretamente a relação entre a variável resposta e as variáveis explicativas. Se o modelo não for adequado, as estimativas dos parâmetros podem ser imprecisas e as previsões podem ser incorretas.

Existem vários testes que podem ser usados para verificar a adequabilidade de modelos de regressão linear. Esses testes incluem:

- **Teste de homocedasticidade**: verifica se a variância dos resíduos é constante ao longo da linha de regressão. (indica algo bom)

- **Teste de normalidade dos resíduos**: verifica se os resíduos seguem uma distribuição normal. (indica algo bom)

- **Teste de autocorrelação**: verifica se existe dependência entre os resíduos. (indica algo ruim)

- **Teste de multicolinearidade**: verifica se existe correlação forte entre as variáveis explicativas. (indica algo ruim)

Se qualquer um desses testes for significativo, isso indica que o modelo não é adequado. Nesse caso, é necessário tomar medidas para melhorar a adequabilidade do modelo, como:

- Transformar as variáveis explicativas ou resposta.

- Reduzir o número de variáveis explicativas.

- Usar um modelo de regressão não linear.


# RESÍDUOS PADRONIZADOS

Os resíduos padronizados são uma ferramenta útil para verificar a adequabilidade de modelos de regressão linear. Eles são calculados dividindo cada resíduo pelo desvio padrão dos resíduos. Isso torna os resíduos independentes da escala da variável resposta.

Os resíduos padronizados também podem ser usados para identificar outliers, que são observações que não seguem o padrão geral dos dados. Outliers podem afetar negativamente a precisão do modelo.

<br>

### **RESÍDUO**
O resíduo é diferença de erro da medida do y observado e o y esperado $(\hat{y})$

$E = Y_{i} - \hat{Y}_{i}$

<br>

### **VARIANÇA RESIDUAL**

Já a variança dos resíduos é a variança das diferenças entre os valores observados da variável resposta e os valores previstos pelo modelo. A variância dos resíduos é uma medida de quanto os resíduos estão espalhados em torno da linha de regressão.

A fórmula é composta dos seguintes termos:

- $s^2$: O desvio padrão dos resíduos.

- $1 - \frac{1}{n}$: Um termo que representa a variância da variável resposta não explicada pelo modelo.

- $\frac{(x - \bar{x})^2}{Sxx}$: Um termo que representa a variância dos resíduos explicada pelo modelo.

###$V (Y_{i} - \hat{Y}_{i}) = s^2 * [1 - \frac{1}{n} - \frac{Sum(x_{i} - \bar{x})^2}{Sxx}]$

<br>

### **RESÍDUOS PADRONIZADOS**

A variância dos resíduos padronizados é usada para verificar a adequabilidade de um modelo de regressão linear, principalmente a suposição de homocedasticidade. A homocedasticidade é a condição em que a variância dos resíduos é constante ao longo da linha de regressão.

Se a variância dos resíduos padronizados for alta, ou seja, não estiver ajustada a reta de regressão, isso pode indicar que o modelo não está descrevendo corretamente a relação entre a variável resposta e as variáveis explicativas. Neste caso, pode ser necessário corrigir a variância dos resíduos usando uma técnica adequada, como a transformação das variáveis.

A variança dos resíduos padronizadas é calculada da seguinte maneira:

- ### $ e_{i}^* = \frac {(Y_{i} - \hat{Y}_{i})} {s^2 \sqrt{ 1 - \frac{1}{n} - \frac{Sum(x_{i} - \bar{x})^2}{Sxx}}}$

# VERIFICANDO A AUSÊNCIA DE COMPORTAMENTO LINEAR

As vezes ao analisar o adequabilidade do modelo verificamos que nem sempre os dados têm um comportamento linear, como por exemplo quando a variança de Y depende de x, o que fazer quando isso ocorre?

<br>

###  Variância dos resíduos não é constante

Quando a variância dos resíduos não é constante, o MMQ pode produzir estimativas dos parâmetros que não são eficientes. Isso significa que as estimativas podem não ser as melhores possíveis, dadas as observações disponíveis.

Para lidar com a heterocedasticidade, existem várias técnicas que podem ser usadas. Uma técnica é transformar as variáveis de modo a tornar a variância dos resíduos constante. Outra técnica é usar um método de estimação que não assume homocedasticidade.

Aqui estão algumas técnicas que podem ser usadas para lidar com a heterocedasticidade:

- **Transformação das variáveis**: As variáveis podem ser transformadas de modo a tornar a variância dos resíduos constante. Por exemplo, as variáveis podem ser transformadas em log ou raízes quadradas.

- **Método dos mínimos quadrados ponderados (MQP)**:  Técnica usada para dar mais importância a algumas observações do que a outras. O MQP é usado quando as observações não são igualmente precisas ou quando a variância dos resíduos não é constante, usando pesos para dar mais importância a algumas observações do que a outras, como peso menor para Yi resultante de x grande e peso maior para Yi resultante de x pequeno.

<br>

###Outliers

Outliers podem afetar a reta ajustada de um modelo de regressão linear. Outliers são observações que estão muito distantes do restante dos dados. Eles podem ser causados por erros de medição, erros de digitação ou outras causas.


<br>

###Erro não distribuição normal


Quando o erro não tem distribuição normal, o modelo de regressão linear pode não ser uma boa aproximação da relação entre a variável resposta e as variáveis explicativas. Isso pode levar a estimativas dos parâmetros que não são precisas.

<br>

###Erro (ε)i depende do tempo

Quando cada observação dos dados denota dependência do tempo e o erro ε depende do tempo, o modelo de regressão linear pode não ser uma boa aproximação da relação entre a variável resposta e as variáveis explicativas. Isso pode levar a estimativas dos parâmetros que não são precisas.

Existem algumas técnicas que podem ser usadas para lidar com essa situação. Uma técnica é usar um modelo de regressão que assume dependência do tempo. Outro método é usar um método de estimação robusto. Aqui estão algumas técnicas que podem ser usadas para lidar com dependência do tempo nos dados:

- Modelos de regressão linear com dependência do tempo: Existem vários modelos de regressão linear que assumem dependência do tempo. Esses modelos podem incluir:

- Regressão linear autorregressiva: Este modelo assume que o erro ε depende dos valores anteriores do erro ε.

- Regressão linear de média móvel: Este modelo assume que o erro ε depende dos valores anteriores da variável resposta.

- Métodos de estimação robustos: Os métodos de estimação robustos são menos sensíveis à dependência do tempo nos dados.











# MODELO NÃO LINEARES

Muitas vezes quando identificamos que nosso modelo não segue a distribuição normal de uma regressão linear, devemos tentar explorar outros modelos probalisticos que envolvem equações não lineares.

Um modelo não-linear é um modelo estatístico no qual a relação entre a variável resposta e as variáveis explicativas não é linear. Em outras palavras, a média da variável resposta não é uma função linear dos parâmetros do modelo.

Alguns exemplos de modelos não-lineares incluem:

- **Regressão logística**: Este modelo é usado para modelar a probabilidade de um evento ocorrer. A probabilidade de um evento ocorrer é expressa como uma função não-linear dos parâmetros do modelo.

- **Regressão polinomial**: Este modelo é usado para modelar a relação entre uma variável resposta contínua e uma ou mais variáveis explicativas contínuas. A relação entre a variável resposta e as variáveis explicativas é expressa como uma função polinomial dos parâmetros do modelo.

- **Regressão exponencial**: Este modelo é usado para modelar a relação entre uma variável resposta contínua e uma ou mais variáveis explicativas contínuas. A relação entre a variável resposta e as variáveis explicativas é expressa como uma função exponencial dos parâmetros do modelo.

- **Regressão logarítmica**: Este modelo é usado para modelar a relação entre uma variável resposta contínua e uma ou mais variáveis explicativas contínuas. A relação entre a variável resposta e as variáveis explicativas é expressa como uma função logarítmica dos parâmetros do modelo.

Os modelos não-lineares são frequentemente usados em situações em que a relação entre a variável resposta e as variáveis explicativas não é linear. Por exemplo, a relação entre a dose de um medicamento e o efeito do medicamento não é linear. Em situações como essa, um modelo linear não seria uma boa aproximação da relação real.

A escolha do modelo não-linear mais adequado depende da forma da relação entre a variável resposta e as variáveis explicativas. Existem vários métodos para escolher o modelo não-linear mais adequado, incluindo métodos gráficos e métodos estatísticos.

# MODELO DE REGRESSÃO EXPONENCIAL


Para linearizar um modelo de regressão exponencial, podemos tomar o logaritmo da equação do modelo. Isso transformará a equação em uma equação linear, que pode ser estimada usando os métodos de regressão linear.

A equação de um modelo de regressão exponencial é a seguinte:

##$y = \alpha * e ^ {\beta*x}$

Onde:

y é a variável resposta

α e β são os parâmetros do modelo

x é a variável explicativa

<br>

Para linearizar temos que converter em logaritmo natural a variável resposta y e usar as entradas convertidas para estimar a regressão junto dos valores originais de x:

$y´ = ln(y)$


<br>

A forma linear da equação será:

$y´ = ln(α) + β * x$

# MODELO DE REGRESSÃO COM POTÊNCIA


Para linearizar um modelo de regressão com função potência, podemos tomar o logaritmo da variável resposta e da variável explicativa. Isso transformará a equação em uma equação linear, que pode ser estimada usando os métodos de regressão linear.

A equação de um modelo de regressão com função potência é a seguinte:

##$y = \alpha * x ^ {\beta}$

Onde:

y é a variável resposta

α e β são os parâmetros do modelo

x é a variável explicativa

Para linearizar temos que converte em logaritmo natural a variável resposta y e a variável explicativa x, e usar as entradas convertidas para estimar a regressão:

$y´ = log(y)$

$x´ = log(x)$

<br>

A forma linear da equação será:

$y´ = log(α) + β * x´$

# REAMOSTRAGEM

A reamostragem é uma técnica que é usada para estimar a incerteza de um modelo probabilístico. Ela funciona gerando amostras aleatórias do conjunto de dados de treinamento e usando essas amostras para estimar os parâmetros do modelo.

Quando estamos trabalhando com modelos probabilísticos, não sabemos com certeza quais são os parâmetros do modelo. Podemos estimar os parâmetros usando os métodos de aprendizado de máquina, mas essas estimativas são apenas uma aproximação da verdade. A reamostragem nos permite estimar a incerteza dessas estimativas.

Existem vários métodos de reamostragem, cada um com suas próprias vantagens e desvantagens. Alguns dos métodos mais comuns incluem:

- Reamostragem Bootstrap: Este método gera amostras aleatórias com reposição do conjunto de dados de treinamento.

- Reamostragem de Jackknife: Este método remove uma observação do conjunto de dados de treinamento de cada vez e estima os parâmetros do modelo usando as observações restantes.

- Reamostragem de Monte Carlo: Este método gera amostras aleatórias sem reposição do conjunto de dados de treinamento.

Aqui estão alguns exemplos de como a reamostragem pode ser usada:

- Para estimar a incerteza de uma previsão: A reamostragem pode ser usada para gerar um conjunto de previsões para um novo dado. Isso pode nos ajudar a entender o intervalo de valores que é provável para a previsão.

- Para comparar modelos concorrentes: A reamostragem pode ser usada para estimar a incerteza dos parâmetros de dois modelos concorrentes. Isso pode nos ajudar a escolher o modelo que é mais provável de ser correto.

- Para realizar inferência causal: A reamostragem pode ser usada para estimar os efeitos causais de uma intervenção. Isso pode ser feito gerando amostras aleatórias de dados sob a suposição de que a intervenção não ocorreu.

## REAMOSTRAGEM BOOTSTRAP

A reamostragem Bootstrap é um método de reamostragem que gera amostras aleatórias com reposição do conjunto de dados de treinamento. Isso significa que cada observação do conjunto de dados pode ser selecionada mais de uma vez.

A reamostragem Bootstrap tem as seguintes características:

- É um método não paramétrico: não faz suposições sobre a distribuição dos dados.

- É um método robusto: é relativamente insensível a violações das suposições do modelo.

- É um método computacionalmente eficiente: pode ser executado rapidamente, mesmo com grandes conjuntos de dados.

A reamostragem Bootstrap é um método de reamostragem que gera amostras aleatórias com reposição do conjunto de dados de treinamento. Isso significa que cada observação do conjunto de dados pode ser selecionada mais de uma vez.

O processo de reamostragem Bootstrap é o seguinte:

Comece com um conjunto de dados de treinamento de tamanho n.
Selecione n observações do conjunto de dados com reposição.
Repita as etapas 2 e 3 B vezes, onde B é o número de amostras Bootstrap desejadas.

Para cada amostra Bootstrap, estime os parâmetros do modelo.
Use as estimativas dos parâmetros das amostras Bootstrap para estimar a incerteza do modelo.

Por exemplo, suponha que temos um conjunto de dados de treinamento com 10 observações. Para gerar uma amostra Bootstrap, selecionaríamos 10 observações do conjunto de dados com reposição. Se selecionarmos a mesma observação mais de uma vez, isso é permitido. Repetiríamos as etapas 2 e 3 100 vezes para gerar 100 amostras Bootstrap.

Para cada amostra Bootstrap, estimaríamos os parâmetros do modelo usando os métodos de aprendizado de máquina. Usando as estimativas dos parâmetros das amostras Bootstrap, poderíamos então estimar a incerteza do modelo.

## REAMOSTRAGEM JACKKNIFE

A reamostragem Jackknife é um método de reamostragem que gera amostras aleatórias sem reposição do conjunto de dados de treinamento. Isso significa que cada observação do conjunto de dados só pode ser selecionada uma vez.

A reamostragem Jackknife tem as seguintes características:

- É um método não paramétrico: não faz suposições sobre a distribuição dos dados.

- É um método robusto: é relativamente insensível a violações das suposições do modelo.

- É um método computacionalmente eficiente: pode ser executado rapidamente, mesmo com grandes conjuntos de dados.

Diferenças entre reamostragem Jackknife e reamostragem Bootstrap:

A principal diferença entre a reamostragem Jackknife e a reamostragem Bootstrap é que a reamostragem Jackknife gera amostras aleatórias sem reposição do conjunto de dados de treinamento, enquanto a reamostragem Bootstrap gera amostras aleatórias com reposição do conjunto de dados de treinamento.

Isso significa que cada observação do conjunto de dados de treinamento só pode ser selecionada uma vez na reamostragem Jackknife, enquanto cada observação do conjunto de dados de treinamento pode ser selecionada mais de uma vez na reamostragem Bootstrap.

Outra diferença entre a reamostragem Jackknife e a reamostragem Bootstrap é que a reamostragem Jackknife é um método de estimativa de ponto, enquanto a reamostragem Bootstrap é um método de intervalo de confiança. Isso significa que a reamostragem Jackknife é usada para estimar o valor do parâmetro do modelo, enquanto a reamostragem Bootstrap é usada para gerar um intervalo de valores que é provável de conter o verdadeiro valor do parâmetro.

# MODELO LOGÍSTICO

O modelo logístico é um modelo estatístico usado para prever a probabilidade de um evento acontecer. Ele é semelhante à regressão linear, exceto que, em vez de um resultado gráfico, a variável de destino é binária; o valor é 1 ou 0.

A regressão logística é usada em uma variedade de aplicações, incluindo:

Predição de inadimplência,
Predição de risco de crédito,
Predição de risco de fraude,
Predição de churn,
Predição de sucesso de marketing e
Predição de diagnóstico médico

A regressão logística é um modelo probabilístico, o que significa que ela fornece uma estimativa da probabilidade de um evento acontecer. Essa estimativa pode ser usada para tomar decisões, como decidir a quem conceder um empréstimo ou a quem oferecer um seguro.

O modelo logístico é baseado na função logística, que é uma função não linear que se aproxima de uma curva sigmoidal. A função logística é usada para transformar os valores das variáveis independentes em valores de probabilidade.

O modelo logisitico é dado pela equação onde y é uma razão de chances de probalidade de um evento acontecer:

###$y = \frac {p(x)}{1-p(x)}$

A equação da função logaritima é dada por:

###$\frac {p(x)}{1-p(x)} = e^{\beta_0 + \beta_1 x}$

O logaritimo da equação da função de chance é uma função linear do preditor, cujo resultado estará no intervalo entre 0 e 1. O Modelo Probalisitico Logistico irá estimar o resultado para a margem mais próxima de acordo com os parâmetros dados.

# MODELO POLINOMIAL

Um modelo de regressão polinomial é um modelo de regressão que usa uma função polinomial para modelar a relação entre uma variável dependente e uma ou mais variáveis independentes. As funções polinomiais são funções do tipo y = ax^n, onde a é um coeficiente e n é o grau do polinômio.

Um modelo de regressão polinomial pode ser usado para modelar uma variedade de relações, incluindo relações lineares, relações não lineares e relações curvilíneas. Por exemplo, um modelo de regressão polinomial pode ser usado para modelar a relação entre o preço de um carro e sua idade, a relação entre o rendimento de um investimento e o tempo, ou a relação entre a temperatura e a altura.

## FUNÇÃO DO MODELO DE REGRESSÃO POLINOMIAL

A função do modelo de regressão polinomial 𝑦 = 𝛽0 + 𝛽1𝑥 + 𝛽2𝑥² + 𝜀 é uma função polinomial de grau 2. Os termos da função são:

𝛽0: é o intercepto da curva de regressão. Representa o valor de 𝑦 quando 𝑥 = 0.
𝛽1: é o coeficiente da variável independente 𝑥. Representa a inclinação da curva de regressão.
𝛽2: é o coeficiente da variável independente 𝑥². Representa a curvatura da curva de regressão.
𝜀: é o termo de erro. Representa a variação não explicada na variável dependente 𝑦.

A função polimional pode assumir k graus, assim também o modelo de regressão a cada grau a função é acrescida do termo + 𝛽𝑘X𝑘 + no k é valor do grau.

O gráfico do modelo de regressão polinomial 𝑦 = 𝛽0 + 𝛽1𝑥 + 𝛽2𝑥² + 𝜀 é uma parábola. A parábola pode ser aberta para cima ou para baixo, dependendo do valor de 𝛽2. Se 𝛽2 for positivo, a parábola será aberta para cima. Se 𝛽2 for negativo, a parábola será aberta para baixo.

O gráfico do modelo de regressão polinomial de grau 3 é uma curva cúbica. A curva cúbica pode ter dois pontos de inflexão, dependendo dos valores dos coeficientes 𝛽1, 𝛽2 e 𝛽3.

O número de pontos de inflexão da curva cúbica representa o número de vezes que a relação entre 𝑦 e 𝑥 muda de concavidade.

## VARIÂNCIA ESTIMADA E COEFICIENTE DE DETERMINAÇÃO

Para modelos lineares a estimativa de σ2 é: 𝝈² = 𝒔² = $\frac {𝑺QE}{n -2}$

Para os modelos polinomiais a variança 𝝈² = 𝒔² = $\frac {𝑺QE}{n - (k +1)}$

O Coeficiente de determinação múltipla: 𝑹² = 1 - $\frac {SQE}{SQT}$

Coeficiente de determinação múltipla ajustado é:

R² ajustado = $1 - \frac {n-1}{n- (k+1)}  * \frac{SQE}{SQT}$ = $\frac {(n-1) R^2 - k}{n-1-k}$

O coeficiente de determinação múltipla ajustado (R² ajustado) é uma medida de ajuste do modelo de regressão polinomial aos valores observados de uma variável aleatória. O R² ajustado varia entre 0 e 1, por vezes sendo expresso em termos percentuais. Nesse caso, expressa a quantidade da variância dos dados que é explicada pelo modelo polinomial. Assim, quanto maior o R² ajustado, mais explicativo é o modelo polinomial, ou seja, melhor ele se ajusta à amostra.

## INTERVALOS ESTATÍSTICOS E PROCEDIMENTOS DE TESTE

- Intervalo de Confiança:
Um IC de 100(1-a)% para b1, o coeficiente de x na regressão polinomial é:

$b1 +/- t. α/2, n-(k+1) * S\beta_1$

Como podemos ver é bem semelhante ao calculo do IC da regressão linear, mudando apenas a formula do grau de liberdade que ao invés de n-k, será n - (k+1).

<br>

- Teste de Hipótese H0:
Um teste de H0 : b1 = b10 baseia-se no valor da estatistica de t igual a:

$t = \frac {\beta_1 - \beta_{10}} {S\beta_{1}}$, se ${\beta_{10}}$ valer zero t será $t = \frac {\beta_1} {S\beta_{1}}$

Se o teste for de desigualde teremos o t >= Tcrit em desfavor de H0 = 0

<br>

- Cálculo do IC do $\hat{y}$ esperado:

$\hat{y} +/- t. α/2, n-(k+1) * S\hat{y}$

<br>

- Cálculo de Intervalo de Previsão:
Um IP de 100(1-a)% para y esperado quando tenho um determinado x é

IP de $\mu_{Y,x^*}$ pode se calcular com $\hat{Y}_1 \pm t_{\frac{\alpha}{2},n-(k+1)} * \sqrt {s^2 + s^2_{\hat{Y}}} $


<br>

##CENTRALIZAÇÃO DE VALORES

- $\beta_0$ é o valor de y para x = 0

- $\beta_1$ é a primeira derivada da função em x = 0 (taxa instântanea de mudança de $\mu_{Y,x^*}$ em x = 0)

- Quanto mais o 𝑥𝑖 se distancia de zero existe a probabilidade de se obter informações imprecisas sobre os parâmetros 𝛽, 𝛽1, 𝛽2, … 𝛽𝑘

# MODELO DE REGRESSÃO MÚLTIPLA

A regressão múltipla é uma técnica estatística que permite estimar a relação linear entre uma variável dependente e múltiplas variáveis independentes. O modelo de regressão múltipla é definido como uma equação linear na qual a variável dependente é a soma dos produtos das variáveis independentes e seus respectivos parâmetros.

Em outras palavras temos uma equação de regressão com mais de uma variável x, onde cada X assume um valor indepedente, ao contrário da regressão polinomial, onde temos vários graus de X com mesmo valor de entrada.

A equação do modelo de regressão múltipla aditiva geral é:
𝐲 = 𝛃𝟎 + 𝛃𝟏X𝟏 + 𝛃𝟐X𝟐 + ⋯ + 𝛃𝐤X𝐤 + 𝛆

## Tipos de Regressão Multipla

Para um conjunto de variáveis 𝐲, 𝐱𝟏 e 𝐱𝟐 um conjunto de quatro modelos possíveis:

### 1 - O modelo de primeira ordem:
𝐲 = 𝛃𝟎 + 𝛃𝟏X𝟏 + 𝛃𝟐X𝟐 + 𝛆

Obs: Duas variáveis independetes x assumindo valores diferente

### 2 - O modelo de segunda ordem sem interação:
𝐲 = 𝛃𝟎 + 𝛃𝟏X𝟏 + 𝛃𝟐X𝟐 + 𝛃𝟐X𝟏² + 𝛃4X𝟐²+ 𝛆

Obs: Duas variáveis independetes x assumindo valores diferente, em polinomio de grau 1 + grau 2

### 3 - O modelo com preditores de primeira ordem e interação:
𝐲 = 𝛃𝟎 + 𝛃𝟏X𝟏 + 𝛃𝟐X𝟐 + 𝛃3X𝟏*X𝟐 + 𝛆

Obs: Duas variáveis independetes x assumindo valores diferente, de grau 1, com uma terceira variável que é o produto entre as duas primeiras

### 4 - O modelo completo de segunda ordem ou modelo quadrático completo:
𝐲 = 𝛃𝟎 + 𝛃𝟏X𝟏 + 𝛃𝟐X𝟐 + 𝛃𝟐X𝟏² + 𝛃4X𝟐²+  𝛃5X𝟏*X𝟐 + 𝛆
Obs: Duas variáveis independetes x assumindo valores diferente, em polinomio de grau 1 + grau 2 + uma quinta variável que é o produto entre as duas primeiras

#MODELOS NÃO LINEARES COM REGRESSÃO MÚLTIPLA

Considerando que estudamos anteriormente os modelos não lineares que são linerarizados para se adequaram a reta de regressão como:

- **Modelos exponenciais**: modelos em que a variável resposta é uma função exponencial das variáveis independentes. Por exemplo, um modelo exponencial pode ser expresso como: $y = α * e ^ {\beta x}$

- **Modelo  com função potência**:  modelo de regressão não linear em que a variável resposta é uma função potência das variáveis independentes. A função potência é uma função não linear que pode ser expressa como: $y = α * x^\beta$

- **Modelo com função logarítmica**: modelo de regressão não linear em que a variável resposta é uma função logarítmica das variáveis independentes. A função logarítmica é uma função não linear que pode ser expressa como: $y = α + \beta * x * log(x)$

- **Modelo com função recíproca**: modelo de regressão não linear em que a variável resposta é uma função recíproca das variáveis independentes. A função recíproca é uma função não linear que pode ser expressa como: $y = α + β * \frac {1}{x}$

- **Modelos Logísticos**: são modelos em que a variável resposta é uma função logística das variáveis independentes. Por exemplo, um modelo logístico pode ser expresso como: $\frac {p(x)}{1- p(x)} = e ^ {\beta_0 + \beta_1 x}$

<br>

Podemos aplicar a estatisitica multivariada a tais modelos, assim como fizmeos no modelo de regressão múltipla, fazendo com que nossos modelos não lineares possam ter mais de uma variável preditora também.

Para issso vamos realizar a transformação de variáveis considerando as regras de cada modelo.

## SELEÇÃO DE VARIÁVEIS

Quando temos um data set com muitas variáveis candidatas a serem preditoras de y, podemos usar algumas regras para definir o melhor modelo aplicado a cada caso.

Geralmente, quanto mais variaveis preditoras temos no modelo, mais o R2 tende a inflar, por causa da inflação de variança, mas esse R2 alto não quer dizer que o coeficiente de determinação é adequado.

Se alguns preditores são adicionados a um determinado modelo, o SQE não pode aumentar, pois um modelo maior não pode explicar menos variação do que um menor, logo o normal é que SQE, geralmente, irá diminuir, talvez por uma pequena
quantidade.

Portanto é Desejável:

⮚ Um modelo envolvendo relativamente poucos preditores.

⮚ Que seja fácil de interpretar e usar.

⮚ Que explique uma quantidade relativamente grande da
variação y observada.

O ideal é trabalhar com o maior valor possível  de R2 para a menor quantidade possível de preditores.

<br>


## QME - QUADRADO MÉDIO DOS ERROS

Outra métrica que podemos usar para quantificar o modelo ideal é o  QME, calculado da seguinte maneira:

QME k = $\frac {SQE_k}{(n - k - 1)}$ sendo k nmr de preditores

Quanto menos preditores maior será o QME k, o objetivo é obter um modelo com o mínimo QME, sendo que podemos usar QME para avaliar o modelo ao invés de R2, pois:

$R2 adj = 1 - \frac {QME}{QMT}$

Sendo $QMT = \frac{SQT}{n-1}$

<br>

## VALOR C k

Outra métrica que podemos usar para quantificar o modelo ideal é o obter o mínimo valor C k em conjunto com o maior valor R2 disponível, sendo C K obtido através de:

$ C_k = \frac {SQE_k}{S^2} + 2 (k + 1) - n$

## Eliminação retroativa (ER) (backward elimination)

O método de eliminação retroativa é um método de seleção de variáveis para modelos de regressão linear. O método começa com um modelo que inclui todas as variáveis independentes. Em seguida, as variáveis são eliminadas uma a uma, começando com a variável que tem o menor impacto no modelo.

O impacto de uma variável no modelo é medido por um teste de hipóteses. Se o teste de hipóteses rejeitar a hipótese nula de que a variável não tem impacto no modelo, a variável é mantida no modelo. Se o teste de hipóteses não rejeitar a hipótese nula, a variável é eliminada do modelo.

Geralmente o teste de hipótese irá verificar se o valor t de cada parametro de modelo é maior que 2 (t critico mais habitual), caso negativo a variável do parametro é retirada do modelo e um novo modelo é gerado com as preditoras restante, e o método é refeito até que todas estejam t > 2.

```python
##@title Gráficos dispersão e resíduos

#Valores previstos e residuos

y_pred=list(res3.predict()) # calcula os valores previsto de y

resi=list(res3.resid) # calcula o residual y - y_pred

# obtem a influencia de cada residuo no resultado
influence = res3.get_influence()

#obter residuos padronizados
stdresid = list(influence.resid_studentized_internal)

# Calcula a razão entre erro observado e  erro padronizado
prop=np.divide(resi,stdresid) #e/e*
```

```python
#Analisando as amostras influentes que podem inflar o modelo

infl = res3.get_influence()
#leverage
print(infl.hat_matrix_diag)

# Como podemos observar não temos nenhuma amostra que ultrapasse muito valor médio com excessão da última que gera uma determinada influencia no modelo
# Por curiosidade esse é resultado do carro Pagani Zonda que tem motorização elevada em relação aos demais carros.
```

```python
# Calcula os valores dos resíduos padronizados (influencias internas)

#residus = res3.resid.as_matrix() # matriz residual já definida acima
leviers = infl.hat_matrix_diag # resultado da matriz

sigma_err = np.sqrt(res3.scale) # erro padrão da regressão

res_stds = stdresid/(sigma_err*np.sqrt(1.0-leviers)) # calcula o denominador do valor padronizado dos resíduos, variança dos resíduos não influenciados

print(res_stds)

# este resultado mostra mais carros que tem uma maior influencia no resultado do modelo em compação a outros que tem valores mais próximos de 0
```

```python
#Limiar


# define um limite para influencia dos pontos de dados e identifica os pontos de dados que ultrapassam esse limite

# O limite é calculado como 2 vezes a razão entre o número de parâmetros (pa) mais 1 e o número de observações (na).

#dimensões
print(dfcars.shape)

#número de instâncias
na = dfcars.shape[0]

# número de variáveis
pa = dfcars.shape[1] - 4

seuil_levier = 2*(pa+1)/na
print(seuil_levier)

# identifica os pontos de dados que ultrapassam o limite.
#A variável atyp_levier é um vetor booleano que indica se cada ponto de dados ultrapassa o limite
atyp_levier = leviers > seuil_levier
print(atyp_levier)

# Retorna que apenas um valor verdadeiro ultrapassou o limite de influencia
```

```python
#Quais carros ultrapassam esse limiar
print(dfcars.index[atyp_levier],leviers[atyp_levier])

# Retorna qual veículo ultrapassou este limite e o valor de influencia
```

```python
#valores de influencia dos dados fornecidos pelo objeto

# Com método
print(infl.resid_studentized_external)

#verificando com a fórmula
res_studs = res_stds*np.sqrt((na-pa-2)/(na-pa-1-res_stds**2))
print(res_studs)
```

```python
#limiares dos resíduos (influencias externas) usando outros parametros

import scipy
seuil_stud = scipy.stats.t.ppf(0.975,df=na-pa-2)
print(seuil_stud)

#detecção - valor absoluto > limite
atyp_stud = np.abs(res_studs) > seuil_stud

#quais são os valores que ultrapassaram o limete?
print(dfcars.index[atyp_stud],res_studs[atyp_stud])

# Como podemos ver ele acrescenta outros valores que não estavam inclusos no teste anterior
```

```python
#Dados de alta influencia considerando ambos critérios

pbm_infl = np.logical_or(atyp_levier,atyp_stud)
print(dfcars.index[pbm_infl])
```

## ANALISANDO MULTICOLINEARIDADE DO MODELO

Verificaremos se há multicolinearidade em nosso modelo através da Regra de Klein e do VIF

```python
# separando somente as variáveis do segundo modelo
dfex = dfcars.drop(columns=["Etype","cemm","cap_vol"]) # dropando as que não serão usadas
dfex.head()
```

```python
# Verificando a correlação entre as variaveis
print(corr)
```

```python
#  regra prática de Klein

# Existe multicolinearidade se a relação mc2 >𝑅2 = 0.966
mc2 = corr**2
print(mc2)
```

```python
# Plotando o mapa de calor do resultado Regra de Klein
sns.heatmap(mc2, cmap='coolwarm', annot=True, fmt=".2f", vmin=-1, vmax=1)
```

```python
# Critério VIF

# Existe multicolinearidade se VIF > 4

vif = np.linalg.inv(mc2)
print(vif)
```

```python
# Plotando gráfico do resultado VIF
sns.heatmap(vif, cmap='coolwarm', annot=True, fmt=".2f", vmin=-4, vmax=4)
```
